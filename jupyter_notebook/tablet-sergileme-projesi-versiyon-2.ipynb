```python
def eksik_deger_tablosu(df): 
    eksik_deger = df.isnull().sum()
    eksik_deger_yuzde = 100 * df.isnull().sum()/len(df)
    eksik_deger_tablo = pd.concat([eksik_deger, eksik_deger_yuzde], axis=1)
    eksik_deger_tablo_son = eksik_deger_tablo.rename(columns = {0 : 'Eksik Deerler', 1 : '% Deeri'})
    return eksik_deger_tablo_son
```

### Model Seim

```python
def train_models(X_train, X_test, y_train, y_test, alg):
    model = alg().fit(X_train, y_train)
    cross_validation_score = cross_val_score(model, X_test, y_test, cv = 10).mean()
    model_name = alg.__name__
    print(model_name, "Cross Validation Score: ",cross_validation_score)
```

```python
models = [LGBMClassifier,
          XGBClassifier,
          GradientBoostingClassifier,
          RandomForestClassifier,
          MLPClassifier,
          GaussianNB,
          AdaBoostClassifier,
          MLPClassifier,
          KNeighborsClassifier]
```
